- **Marital Status:** Marital status (Single, Married, Divorced).

- **NumCompaniesWorked:** Number of companies the employee has worked for (integer).

- **Over18:** Whether the employee is over 18 (Yes/No).

- **PercentSalaryHike:** Percentage increase in salary (float).

- **PerformanceRating:** Performance rating (float, probably a categorical encoding).

#### PANDAS WITH FUNCTIONS
- Let's assume the daily rate has increased by 10%

```python
hr_df2 = pd.read_csv('Human_Resources.csv')
hr_df2
```

```python
hr_df2['ten_per'] = hr_df2['DailyRate'] * 0.10
print(hr_df2['ten_per'])

# This helps in understanding the value of a 10% increase in daily rates for each employee.
```

```python
print ("the daily rate before the 10 increase: ",hr_df2['DailyRate'])

hr_df2['DailyRate'] = hr_df2['DailyRate'] + hr_df2['ten_per']

print ("\nthe daily rate after the 10 increase: ",hr_df2['DailyRate'])

# It shows how the daily rates change after a 10% increase, which can be used for financial planning and budgeting.
```

```python
hr_df2['Net_Worth'] = hr_df2['MonthlyIncome'] * 12

# This gives a yearly financial perspective of each employee's earnings.

def increase_net_worth(df, net_worth_column, increase_percent= 0.10):

    df[net_worth_column] = df[net_worth_column] * (1 + increase_percent)
    return df

# This provides a flexible way to apply percentage-based increases to any financial column in the DataFrame.

hr_df2 = increase_net_worth(hr_df2, 'Net_Worth')

print(hr_df2['Net_Worth'])

# It shows how the annual net worth of employees changes with a 10% increase, which is useful for scenarios like salary negotiations or financial projections.
```

##### The summary:
1. The calculations show how a 10% increase in daily rate and net worth affects employees differently based on their initial salaries.
2. The column net_worth reflects each employee‚Äôs annual net worth after the 10% increase in their monthly income. It highlights the annualized effect of the salary adjustments. It is useful for understanding long-term financial impacts and for making compensation-related decisions.
